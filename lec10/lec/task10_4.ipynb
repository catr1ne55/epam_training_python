### 1. Постройте график числа золотых медалей, выигранных США мужчинами и женщинами в атлетике?

```python
import seaborn as sns
```

```python
# спортсмены из США, выигравшие золотые медали в атлетике
persons = oo[(oo.Medal == 'Gold') & (oo.Discipline == 'Athletics') & (oo.NOC == 'USA')]
# строим график
sns.countplot(x='Edition',data=persons, hue='Gender');
```

### 2. Постройте график 5 атлетов, которые выиграли большинство золотых медалей. 

```python
# сначала вычислим пятерых лучших атлетов по количеству золотых медалей
only_gold = oo[oo.Medal == 'Gold']
only_gold.Athlete.value_counts()[:5]
```

#### Вариант построения графиков отдельно для каждого спортсмена:

```python
only_gold.loc[oo.Athlete == 'PHELPS, Michael'].groupby('Edition').agg({'Medal' : ['count']}).plot(kind='bar');
```

```python
only_gold.loc[oo.Athlete == 'LATYNINA, Larisa'].groupby('Edition').agg({'Medal' : ['count']}).plot(kind='bar');
```

```python
only_gold.loc[oo.Athlete == 'SPITZ, Mark'].groupby('Edition').agg({'Medal' : ['count']}).plot(kind='bar');
```

```python
only_gold.loc[oo.Athlete == 'NURMI, Paavo'].groupby('Edition').agg({'Medal' : ['count']}).plot(kind='bar');
```

```python
only_gold.loc[oo.Athlete == 'LEWIS, Carl'].groupby('Edition').agg({'Medal' : ['count']}).plot(kind='bar');
```

#### Все спортсмены на одном графике:

```python
top5_gold = only_gold[(only_gold.Athlete == 'PHELPS, Michael') | (only_gold.Athlete == 'LATYNINA, Larisa') | (only_gold.Athlete == 'SPITZ, Mark') | (only_gold.Athlete == 'NURMI, Paavo') | (only_gold.Athlete == 'LEWIS, Carl')]
sns.countplot(x='Edition',data=top5_gold, hue='Athlete');
```
