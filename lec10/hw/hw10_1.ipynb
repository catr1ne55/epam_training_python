### Постройте таблицу в которой по годам всех олимпиад покажите топовых атлетов США(1 атлет на год) по общему количеству медалей? Включите дисциплину атлета.

```python
usa = oo[oo.NOC == 'USA']
aggregations = {'Medal': 'count'}
# построим таблицу, где у каждого атлета указано количество заработанных медалей за олимпиаду
groups = usa.groupby(['Edition', 'Athlete', 'Discipline']).agg(aggregations).reset_index()

def best_result(groups):
    max_number_medals = max(groups.Medal)
    for i in groups.index:
        if groups.Medal[i] == max_number_medals:
            return groups.loc[i]
        
# выберем по одному лучшему атлету за год        
top1_for_year = groups.groupby('Edition').apply(best_result).drop(columns='Edition')
top1_for_year
```
