### 1. Используя  groupby(), постройте график  - число всех медалей, выигранных на каждой олимпиаде.

```python
oo.groupby('Edition').agg({'Medal' : ['count']}).plot(kind='bar', title='Число всех медалей, выигранных на каждой олимпиаде');
```

### 2. Создайте список, показывающий число всех медалей, выигранных каждой страной в течение всей истории олимпийских игр. Для каждой страны необходимо показать год первой и последней заработанной медали.

```python
# создадим список, каждый элемент которого содержит название страны, год первой заработанной медали,
# год последней заработанной медали и общее количество медалей за все время
medals_count = list(oo.groupby(['NOC']).agg({'Edition' :['min','max'], 'Medal' : ['count']}).to_dict().values())
list_medals_count = []
for key in medals_count[0].keys() & medals_count[1].keys() & medals_count[2].keys():
    list_medals_count.append((key, medals_count[0][key], medals_count[1][key], medals_count[2][key]))
print(sorted(list_medals_count))
```
