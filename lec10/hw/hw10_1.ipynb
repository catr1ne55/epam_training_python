### Постройте таблицу в которой по годам всех олимпиад покажите топовых атлетов США(1 атлет на год) по общему количеству медалей? Включите дисциплину атлета.

```python
# выберем только атлетов из США
usa = oo[oo.NOC == 'USA']
aggregations = {'Medal': 'count'}
# построим таблицу, где у каждого атлета указано количество заработанных золотых, серебряных и бронзовых
# медалей за олимпиаду соответственно
groups = usa.groupby(['Edition', 'Athlete', 'Discipline', 'Medal']).agg(aggregations).unstack('Medal', fill_value=0)
groups.columns = groups.columns.droplevel()
# добавим колонку с общим количеством заработанных медалей для каждого спортсмена
groups['Total'] =  groups['Gold'] + groups['Silver'] + groups['Bronze']
groups = groups.reset_index()
```

```python
# метод, позволяющий вычислить максимальное количество медалей, заработанных за олимпиаду,
# а также атлета, соответствующего этому количеству медалей
def best_result(groups):
    max_number_medals = max(groups.Total)
    for i in groups.index:
        if groups.Total[i] == max_number_medals:
            return groups.loc[i]
        
# выберем по одному лучшему атлету за год, используя метод, описанный выше        
top1_for_year = groups.groupby('Edition').apply(best_result).drop(columns='Edition')
top1_for_year.columns.name = 'Idx'
top1_for_year = top1_for_year.reset_index()
# в данном случае мы учитываем только одного атлета с максимальным количеством медалей, который попадается первым,
# если таковых несколько за год
top1_for_year
```
